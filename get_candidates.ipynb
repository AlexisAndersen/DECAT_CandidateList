```python
## Open a html file of the candidate webpage with initial cuts ##

with open("candidate_list.html", "r") as file:
    html = file.read()

soup = BeautifulSoup(html)
```

```python
## Define a function that creates a csv of all the candidates on the webap ##

def find_candidates(soup):
    rows = soup.body.find_all("tr")
    print(len(rows))
    candidates = []
    # print(rows[2])
    candidates = []
    for i in range(2, len(rows)-2):
        # x1 = float(row[i].attrs["y"])
        s = BeautifulSoup(str(rows[i]))
        a = s.find("a")
        candidates.append(a.text)
    df = pd.DataFrame({"Candidates": candidates})
    df.to_csv("candidate_list.csv")
    
    return df
    
find_candidates(soup)
```
